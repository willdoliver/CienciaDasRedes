```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd

allCategories = pd.read_csv("../categories.csv", index_col=0)
G = nx.read_gml("../GraphMissingEdges.gml")
```

```python
degree = nx.degree_centrality(G)
sort_orders = sorted(degree.items(), key=lambda degree: degree[1], reverse=True)
# print(sort_orders)
top50 = sort_orders[:50]

dfDegree = pd.DataFrame(columns=['key','item','name','categories'])

print("Centralidade de grau")
for key,item in top50:
    cats = G.nodes[key]['categories'].split(",")
    categoriesString = []
    for cat in cats:
        categoriesString.append(allCategories.loc[int(cat)]['names'])

    dfDegree.loc[key] = pd.Series({'key':key ,'item':round(item,6), 'name':G.nodes[key]['name'], 'categories':", ".join(categoriesString)})

print(dfDegree)
# print(dfDegree.info())
# print(dfDegree.index)
dfDegree.to_csv('degreeCentrality.csv', sep=',', encoding='utf-8', index=False)
```
