```python
# 2- Crie uma rede aleatória Erdos-Renyi com N=3000 e com probabilidade p=10-3 de dois nós aleatórios se conectarem
n = 3000  # Nodes
p = 0.001 # Probability
G = nx.gnp_random_graph(n, p)
nx.draw(G)
```

```python
# a) Qual o número de links esperado? Isso vai de acordo com o que foi observado na rede criada?
```

```python
# Real network <k> << N

linksNumberMean = (n*(n-1))/2
print("linksNumberMean: "+str(linksNumberMean))
```

Através do grafo desenhado, por ser muito grande fica mais difícil a visualização das arestas, mas no gephi é possível ver que no centro há centenas de nós com muitas conexões, o que pode justificar o valor de links esperados mas o grafo realmente parece estar com menos nós do que a média, pois os nós desconectados ao redor formam um anél e apesar de estarem espalhados possuem um número espressivo de nós

```python
# b) Em qual regime essa rede se encontra, subcritical, critical, supercritical ou conected? Justifique.
```

```python
dicDegree = dict(G.degree())
valores = sorted(set(dicDegree.values()))
hist = [list(dicDegree.values()).count(x) for x in valores]
plt.plot(valores, hist, 'ro-') 
plt.xlabel('Grau')
plt.ylabel('Numero de nós')
```

```python
degreeNumberMean = p*(n-1)
print("degreeNumberMean: " + str(degreeNumberMean))

if degreeNumberMean < 1:
    print("Subcritical")
elif degreeNumberMean == 1:
    print("Critical")
elif degreeNumberMean < (n/10): # N > 1 and N < 300 (3000*0.1)
    print("Supercritical")
else: # degreeNumberMean >> N
    print("Connected")
```

Esta rede é supercrítica, pois possui um número médio de 2.9 arestas por nó, o que gera nós no centro com muitas conexões e nós poucos conectados não extremidades, podendo ocorrer de haver nós sem conexões, como esse número médio está próximo de 1, a rede pode vir a ser mais próxima de crítica, pois 2.9 é um valor de rede baixo para supercrítica

```python
# b) Compute a distribuição de graus para duas redes aleatórias com o mesmo número de nós e não direcionadas,
# mas uma com p=0,001 e outra com p =0,01.

n1 = 500
n2 = 1000
n3 = 2000

p1 = 0.001
# 10 times more links probability
p2 = 0.01

P1N1 = nx.gnp_random_graph(n1, p1)
P1N2 = nx.gnp_random_graph(n2, p1)
P1N3 = nx.gnp_random_graph(n3, p1)
```

```python
dicDegree = dict(P1N1.degree())
valores = sorted(set(dicDegree.values()))
hist = [list(dicDegree.values()).count(x) for x in valores]
plt.plot(valores, hist, 'ro-', color="green") 
plt.xlabel('Grau')
plt.ylabel('Numero de nós')

dicDegree = dict(P1N2.degree())
valores = sorted(set(dicDegree.values()))
hist = [list(dicDegree.values()).count(x) for x in valores]
plt.plot(valores, hist, 'ro-', color="blue") 
plt.xlabel('Grau')
plt.ylabel('Numero de nós')

dicDegree = dict(P1N3.degree())
valores = sorted(set(dicDegree.values()))
hist = [list(dicDegree.values()).count(x) for x in valores]
plt.plot(valores, hist, 'ro-', color="red") 
plt.xlabel('Grau')
plt.ylabel('Numero de nós')
```

```python
P2N1 = nx.gnp_random_graph(n1, p2)
P2N2 = nx.gnp_random_graph(n2, p2)
P2N3 = nx.gnp_random_graph(n3, p2)

dicDegree = dict(P2N1.degree())
valores = sorted(set(dicDegree.values()))
hist = [list(dicDegree.values()).count(x) for x in valores]
plt.plot(valores, hist, 'ro-', color="green") 
plt.xlabel('Grau')
plt.ylabel('Numero de nós')

dicDegree = dict(P2N2.degree())
valores = sorted(set(dicDegree.values()))
hist = [list(dicDegree.values()).count(x) for x in valores]
plt.plot(valores, hist, 'ro-', color="blue") 
plt.xlabel('Grau')
plt.ylabel('Numero de nós')

dicDegree = dict(P2N3.degree())
valores = sorted(set(dicDegree.values()))
hist = [list(dicDegree.values()).count(x) for x in valores]
plt.plot(valores, hist, 'ro-', color="red") 
plt.xlabel('Grau')
plt.ylabel('Numero de nós')
```
